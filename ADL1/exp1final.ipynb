```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python

student_marks = np.random.randint(50, 101, size=50)
```

```python
mean_marks = np.mean(student_marks)
median_marks = np.median(student_marks)
std_deviation = np.std(student_marks)
variance = np.var(student_marks)
max_marks = np.max(student_marks)
min_marks = np.min(student_marks)
```

```python
print("Student Marks Statistical Analysis:")
print(f"Total Students: {len(student_marks)}")
print(f"Mean Marks: {mean_marks:.2f}")
print(f"Median Marks: {median_marks:.2f}")
print(f"Standard Deviation: {std_deviation:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Maximum Marks: {max_marks}")
print(f"Minimum Marks: {min_marks}")
```

```python
plt.figure(figsize=(10, 6))
plt.hist(student_marks, bins=10, edgecolor='black')
plt.title('Distribution of Student Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python

percentiles = [25, 50, 75]
for p in percentiles:
    print(f"{p}th Percentile: {np.percentile(student_marks, p):.2f}")
```
