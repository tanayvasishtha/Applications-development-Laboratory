```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
math_marks = np.random.normal(loc=70, scale=10, size=100)
science_marks = np.random.normal(loc=65, scale=12, size=100)
```

```python
plt.figure(figsize=(12, 4))
```

```python
plt.subplot(1, 2, 1)
plt.hist(math_marks, bins=10, edgecolor='black')
plt.title('Distribution of Math Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
```

```python
plt.subplot(1, 2, 2)
plt.scatter(math_marks, science_marks)
plt.title('Math Marks vs Science Marks')
plt.xlabel('Math Marks')
plt.ylabel('Science Marks')
```

```python
plt.tight_layout()
plt.show()
```
