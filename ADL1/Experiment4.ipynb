```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [50000, 55000, 52000, 60000, 65000, 63000]
}
sales_df = pd.DataFrame(data)
```

```python
sales_array = np.array(sales_df['Sales'])
growth_rates = np.zeros_like(sales_array)
growth_rates[1:] = (sales_array[1:] - sales_array[:-1]) / sales_array[:-1] * 100
```

```python

sales_df['Growth Rate'] = growth_rates
```

```python

plt.figure(figsize=(12, 4))
```

```python
plt.subplot(1, 2, 1)
plt.plot(sales_df['Month'], sales_df['Sales'], marker='o')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
```

```python
plt.subplot(1, 2, 2)
plt.bar(sales_df['Month'][1:], sales_df['Growth Rate'][1:])
plt.title('Monthly Sales Growth Rates')
plt.xlabel('Month')
plt.ylabel('Growth Rate (%)')
```

```python
plt.tight_layout()
plt.show()
```
