```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
temperatures = np.random.normal(loc=20, scale=5, size=len(dates))
```

```python
temp_df = pd.DataFrame({
    'Date': dates,
    'Temperature': temperatures
})
temp_df.set_index('Date', inplace=True)
```

```python
rolling_mean_7_days = temp_df['Temperature'].rolling(window=7).mean()
rolling_mean_30_days = temp_df['Temperature'].rolling(window=30).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(temp_df.index, temp_df['Temperature'], label='Daily Temperature', alpha=0.5)
plt.plot(temp_df.index, rolling_mean_7_days, label='7-Day Rolling Average', color='red')
plt.plot(temp_df.index, rolling_mean_30_days, label='30-Day Rolling Average', color='green')

plt.title('Daily Temperature and Rolling Averages')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.tight_layout()
plt.show()
```
